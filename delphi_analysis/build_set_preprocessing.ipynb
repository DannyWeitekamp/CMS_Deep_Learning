```python
if __package__ is None:
    import sys, os
    sys.path.append(os.path.realpath("/data/shared/Software/"))
from CMS_SURF_2016.utils.preprocessing import *

#Set up all of our preprocessing parameters
observ_types = ['E/c', 'Px', 'Py', 'Pz', 'Charge', "PT_ET", "Eta", "Phi", "Dxy_Ehad_Eem"]

#Define the maxes as -1 to indicated that we will procedurally find them with resolveProfileMaxes
object_profiles = [ObjectProfile("Electron",          -1, sort_columns=["PT_ET"], sort_ascending=False),
                    ObjectProfile("MuonTight",        -1, sort_columns=["PT_ET"], sort_ascending=False),
                    ObjectProfile("Photon",           -1, sort_columns=["PT_ET"], sort_ascending=False),
                    ObjectProfile("MissingET",        -1, sort_columns=["PT_ET"], sort_ascending=False),
                    ObjectProfile("EFlowPhoton",      -1, sort_columns=["PT_ET"], sort_ascending=False),  
                    ObjectProfile("EFlowNeutralHadron",-1, sort_columns=["PT_ET"], sort_ascending=False), 
                    ObjectProfile("EFlowTrack",       -1, sort_columns=["PT_ET"], sort_ascending=False)] 


label_dir_pairs = \
            [   ("ttbar", "/data/shared/Delphes/ttbar_lepFilter_13TeV/pandas_unjoined/"),
                ("wjet", "/data/shared/Delphes/wjets_lepFilter_13TeV/pandas_unjoined/"),
                ("qcd", "/data/shared/Delphes/qcd_lepFilter_13TeV/pandas_unjoined/")
            ]

#Resolve the profile maxes, we use 1.0 so we do not pad any more than necessary
resolveProfileMaxes(object_profiles, label_dir_pairs,padding_multiplier = 1.0)
```

```python
preprocessed_dir = "/data/shared/Delphes/preprocessed/"
start = 0
end = maxMutualLength(label_dir_pairs, object_profiles)
stride = 1000
for proc_start in range(start, end, stride):
    proc_num = min(stride, end-proc_start)
    print(proc_start, proc_num)
    X,Y = preprocessFromPandas_label_dir_pairs(
            label_dir_pairs,
            proc_start,
            proc_num,
            object_profiles,
            observ_types
    )
    XY_to_pickle(X,Y, preprocessed_dir+'data_'+str(proc_start/1000) + 'k')
        
        
```
